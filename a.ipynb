```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
import plotly.express as px
```

```python
# convert into numeric, removing commas and coercing errors to NaN
space_df['Price'] = pd.to_numeric(space_df['Price'].astype(str).str.replace(',', ''), errors='coerce')
```

```python
# fill NaN values with 0 and convert to float
space_df['Price'] = space_df['Price'].fillna(0).astype(float)
```

```python
# extract country from Location column
space_df['countries'] = space_df['Location'].str.split(',').str[-1].str.strip()
```

```python
# take states out of USA
usa_states = ['CA', 'FL', 'TX', 'Texas', 'New Mexico', 'Virginia']
space_df['countries'] = space_df['countries'].replace(usa_states, 'USA')
```

```python
# extract year from Datum column using regex
def extract_year(text):
    if pd.isna(text):
        return None
    match = re.search(r'(19\d{2}|20\d{2})', str(text))
    return int(match.group(1)) if match else None

space_df['date'] = space_df['Datum'].apply(extract_year)
```

```python
# convert date to Int64 
space_df['date'] = space_df['date'].astype('Int64')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Mapping Function: Isolates NASA and groups everything else American as 'USA'
def map_entities(row):
    company = str(row['Company Name']).strip()
    country = str(row['countries']).strip()
    
    # Priority 1: Keep NASA separate
    if 'NASA' in company:
        return 'NASA'
    
    # Priority 2: Label all other US launches simply as 'USA'
    if country == 'USA':
        return 'USA'
    
    # Priority 3: International Rivals
    if country == 'China':
        return 'China'
    if country in ['Russia', 'USSR', 'Russia/USSR']:
        return 'Russia/USSR'
    
    return None

# 2. Filter data (1960–2025)
era_df = space_df[(space_df['date'] >= 1960) & (space_df['date'] <= 2025)].copy()

# 3. Apply the categories
era_df['Category'] = era_df.apply(map_entities, axis=1)
df_filtered = era_df[era_df['Category'].notna()]

# 4. Pivot for plotting
pivot_df = df_filtered.groupby(['date', 'Category']).size().unstack(fill_value=0)

# 5. Create the Visualization
plt.style.use('ggplot')
fig, ax = plt.subplots(figsize=(14, 7))

# Custom Colors: NASA (Navy), USA (Blue), Russia (Red), China (Gold)
colors_map = {
    'NASA': '#2c3e50',           
    'USA': '#3498db', 
    'Russia/USSR': '#e74c3c',    
    'China': '#f1c40f'           
}

# Ensure specific order in the legend
order = ['NASA', 'USA', 'Russia/USSR', 'China']
plot_cols = [c for c in order if c in pivot_df.columns]

pivot_df[plot_cols].plot(kind='line', marker='o', markersize=4, linewidth=2, 
                         ax=ax, color=[colors_map[col] for col in plot_cols])

# 6. Formatting
ax.set_title('Global Launch Trends: NASA, USA, and International Rivals (1960–2025)', fontsize=16, fontweight='bold')
ax.set_ylabel('Annual Launch Count', fontsize=12)
ax.set_xlabel('Year', fontsize=12)
ax.set_xticks(range(1960, 2026, 5))
ax.grid(True, linestyle='--', alpha=0.6)
ax.legend(title="Entity / Country")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Reuse the Mapping Logic to identify our 5 focus groups
def map_to_focus_groups(row):
    company = str(row['Company Name']).strip()
    country = str(row['countries']).strip()
    
    if country == 'China': return 'China'
    if country in ['Russia', 'USSR', 'Russia/USSR']: return 'Russia'
    if 'SpaceX' in company: return 'SpaceX'
    if 'NASA' in company: return 'NASA'
    if 'ULA' in company or 'United Launch' in company: return 'ULA'
    return None

# 2. Filter for the modern era (2000–2026)
df_heat = space_df[(space_df['date'] >= 2000) & (space_df['date'] <= 2026)].copy()
df_heat['Group'] = df_heat.apply(map_to_focus_groups, axis=1)

# 3. Create the Pivot Table (Rows = Entities, Columns = Years)
# We count the number of missions for each entity per year
heatmap_data = df_heat[df_heat['Group'].notna()].pivot_table(
    index='Group', 
    columns='date', 
    values='Status Mission', 
    aggfunc='count'
).fillna(0)

# 4. Create the Visualization
plt.figure(figsize=(18, 8))
sns.set_theme(style="white")

# Plotting with the "magma" palette (dark to bright) 
# to emphasize high launch counts
ax = sns.heatmap(
    heatmap_data, 
    annot=True,      # Show the actual launch counts
    fmt=".0f",       # No decimals
    cmap="magma",    # Dark purple/black for low, bright yellow/white for high
    linewidths=.5, 
    cbar_kws={'label': 'Annual Launch Count'}
)

# 5. Formatting
plt.title('Launch Density: Key Global & Commercial Players (2000–2026)', fontsize=18, fontweight='bold', pad=20)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Entity', fontsize=14)

plt.tight_layout()
plt.show()
```
